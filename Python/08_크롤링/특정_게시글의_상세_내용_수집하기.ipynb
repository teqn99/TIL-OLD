```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
import xlwt

f_name = input('검색 결과를 저장할 파일경로와 이름을 지정하세오(예: test.txt): ')
fc_name = input('검색 결과를 저장할 csv 파일경로와 이름을 지정하세요(예: test.csv): ')
fx_name = input('검색 결과를 저장할 xls 파일경로와 이름을 지정하세요(예: test.xls): ')
```

```python
# Step 1. 크롬 드라이버를 사용해서 웹 브라우저를 실행합니다.
path = './chromedriver'  # 크롬 드라이버의 경로 입력
driver = webdriver.Chrome(path)

driver.get('https://korean.visitkorea.or.kr/detail/rem_detail.html?cotid=be3db10c-b642-409c-81cc-c4cdecb5bd8b&temp=')  # 한국관광공사 사이트 링크 사용
time.sleep(2)  # 창이 모두 열릴 때 까지 2초 기다리기
```

```python
# 소주제 1: 본문 내용 출력하기
# Step 2. 현재 페이지에 있는 텍스트 내용을 화면에 출력하기

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
content_list = soup.find_all('div', 'txt_p')
print(content_list)
```

```python
# 본문 내용 수집하기
contents2 = []

for i in content_list:
    print('\n')
    cont = i.get_text()
    contents2.append(cont)
    print(contents2)
```

```python
# 태그 정보 찾기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
tag_list = soup.find_all('ul', 'tag clfix')

tags2 = []
for i in tag_list:
    print('\n')
    tag = i.get_text()
    tags2.append(tag)
print(tags2)
```

```python
# 소주제 2: 출력된 결과를 파일로 저장하기

# 출력 결과를 txt 파일로 저장하기
f = open(f_name, 'a', encoding='UTF-8')
f.write(str(contents2))
f.write(str(tags2))
f.close()
print('txt 파일 저장 경로: %s' % f_name)

# 출력 결과를 표(데이터 프레임)형태로 만들기
import pandas as pd

no2 = 1

korea_trip = pd.DataFrame()
korea_trip['번호'] = no2
korea_trip['내용'] = pd.Series(contents2)
korea_trip['태그'] = pd.Series(tags2)

# csv 형태로 저장하기
korea_trip.to_csv(fc_name, encoding='UTF-8-sig')
print('csv 파일 저장 경로: %s' % fc_name)

# 엑셀 형태로 저장하기
korea_trip.to_excel(fx_name)
print('xls 파일 저장 경로: %s' % fx_name)
```
